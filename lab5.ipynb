```python
import pandas as pd

# Step 1: Create Customers dataset
customers = {
    "customer_id": [1, 2, 3, 4, 5],
    "name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "city": ["Delhi", "Mumbai", "Delhi", "Chennai", "Mumbai"]
}
df_customers = pd.DataFrame(customers)

print("🔹 Customers Dataset:")
print(df_customers)

# Step 2: Create Transactions dataset
transactions = {
    "transaction_id": [101, 102, 103, 104, 105, 106, 107, 108],
    "customer_id": [1, 2, 1, 3, 4, 5, 2, 3],
    "amount": [500, 700, 200, 1200, 1500, 800, 300, 1000]
}
df_transactions = pd.DataFrame(transactions)

print("\n🔹 Transactions Dataset:")
print(df_transactions)

# Step 3: Merge datasets using customer_id
df_merged = pd.merge(df_transactions, df_customers, on="customer_id", how="inner")

print("\n✅ Merged Dataset:")
print(df_merged)

# Step 4: Calculate total spend per customer
customer_spend = df_merged.groupby(["customer_id", "name"])["amount"].sum().reset_index()
customer_spend.rename(columns={"amount": "total_spend"}, inplace=True)

print("\n💰 Total Spend per Customer:")
print(customer_spend)

# Step 5: Group by city to analyze spending
city_spend = df_merged.groupby("city")["amount"].sum().reset_index()
city_spend.rename(columns={"amount": "city_total_spend"}, inplace=True)

print("\n🏙️ City-wise Spending:")
print(city_spend)

# Step 6: Identify top spender
top_spender = customer_spend.loc[customer_spend["total_spend"].idxmax()]

print("\n👑 Top Spender:")
print(top_spender)
```

```python
import pandas as pd

# Step 1: Create Patients dataset
patients = {
    "patient_id": [1, 2, 3, 4, 5],
    "name": ["Ravi", "Anita", "Kiran", "Meena", "Arjun"],
    "age": [34, 45, 29, 60, 52]
}
df_patients = pd.DataFrame(patients)

print("🔹 Patients Dataset:")
print(df_patients)

# Step 2: Create Appointments dataset
appointments = {
    "appointment_id": [101, 102, 103, 104, 105, 106, 107, 108, 109],
    "patient_id": [1, 2, 1, 3, 4, 2, 5, 3, 1],
    "doctor": ["Dr. Sharma", "Dr. Gupta", "Dr. Sharma", "Dr. Iyer",
               "Dr. Gupta", "Dr. Sharma", "Dr. Iyer", "Dr. Gupta", "Dr. Sharma"],
    "date": ["2025-09-01", "2025-09-02", "2025-09-05", "2025-09-06",
             "2025-09-07", "2025-09-08", "2025-09-09", "2025-09-10", "2025-09-12"]
}
df_appointments = pd.DataFrame(appointments)

print("\n🔹 Appointments Dataset:")
print(df_appointments)

# Step 3: Merge datasets using patient_id
df_merged = pd.merge(df_appointments, df_patients, on="patient_id", how="inner")

print("\n✅ Merged Dataset:")
print(df_merged)

# Step 4: Count appointments per patient
patient_visits = df_merged.groupby(["patient_id", "name"])["appointment_id"].count().reset_index()
patient_visits.rename(columns={"appointment_id": "total_visits"}, inplace=True)

print("\n👩‍⚕️ Appointments per Patient:")
print(patient_visits)

# Step 5: Group by doctor to compute workload
doctor_workload = df_merged.groupby("doctor")["appointment_id"].count().reset_index()
doctor_workload.rename(columns={"appointment_id": "total_appointments"}, inplace=True)

print("\n🩺 Doctor Workload:")
print(doctor_workload)

# Step 6: Find patient with maximum visits
top_patient = patient_visits.loc[patient_visits["total_visits"].idxmax()]

print("\n👑 Patient with Maximum Visits:")
print(top_patient)
```
