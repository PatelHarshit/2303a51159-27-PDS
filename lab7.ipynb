```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (assuming file name is 'co2_emissions.csv')
# The CSV should have columns: "Year", "CO2_Emissions"
data = pd.read_csv("co2_emissions.csv")

# Display first few rows
print(data.head())

# Plotting Line Chart
plt.figure(figsize=(10,6))
plt.plot(data["Year"], data["CO2_Emissions"], marker='o', linestyle='-', linewidth=2)

# Chart details
plt.title("Global CO2 Emissions Over Time", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("CO2 Emissions (Million Metric Tons)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (assuming file name is 'student_scores.csv')
# The CSV should have columns: "Exam", "Score"
data = pd.read_csv("student_scores.csv")

# Display first few rows
print(data.head())

# Plotting Line Chart
plt.figure(figsize=(10,6))
plt.plot(data["Exam"], data["Score"], marker='o', linestyle='-', linewidth=2, color="blue")

# Chart details
plt.title("Student Performance Over Exams", fontsize=14)
plt.xlabel("Exam Number", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()

# Show plot
plt.show()
```
